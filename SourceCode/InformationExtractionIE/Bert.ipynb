```python
import ast
from datasets import Features, Sequence, Value, ClassLabel

label_names = [
    "O",
    "B-PER",
    "I-PER",
    "B-ORG",
    "I-ORG",
    "B-LOC",
    "I-LOC",
    "B-DATE",
    "I-DATE",
    "B-PRICE",
    "I-PRICE",
    "B-PRODUCT",
    "I-PRODUCT",
    "B-EVENT",
    "I-EVENT",
    "B-ORDER_ID",
    "I-ORDER_ID",
]


def preprocess_data(example):
    example['tokens'] = ast.literal_eval(example['tokens'])
    example['ner_tags'] = ast.literal_eval(example['ner_tags'])
    return example

processed_datasets = dataset_dict.map(preprocess_data)

features = Features({
    'tokens': Sequence(Value('string')),
    'ner_tags': Sequence(ClassLabel(names=label_names))
})

def update_labels(example):
    example['ner_tags'] = [int(tag) for tag in example['ner_tags']]
    return example

raw_datasets = processed_datasets.map(update_labels, features=features)
```

```python
def align_labels_with_tokens(labels, word_ids):
    """
    Căn chỉnh nhãn với các token đã được tokenized, xử lý cả các token đặc biệt và subword.
    """
    new_labels = []
    current_word = None

    for word_id in word_ids:
        # Token đặc biệt hoặc ngoài phạm vi word_id
        if word_id is None or word_id >= len(labels):
            new_labels.append(-100)
            continue

        # Token đầu tiên của một từ mới
        if word_id != current_word:
            current_word = word_id
            new_labels.append(labels[word_id])
        else:
            # Token thuộc subword: chuyển B-* thành I-*
            new_labels.append(labels[word_id] + 1 if labels[word_id] % 2 == 1 else labels[word_id])

    return new_labels

```

```python
def tokenize_and_align_labels(examples):
    # Tokenize inputs
    tokenized_inputs = tokenizer(
        examples["tokens"],
        truncation=True,
        is_split_into_words=True,  # Ensures tokenizer recognizes word-level input
    )

    # Align labels with tokenized inputs
    aligned_labels = []
    for batch_index in range(len(examples["tokens"])):
        word_ids = tokenized_inputs.word_ids(batch_index=batch_index)  # Get word IDs
        labels = examples["ner_tags"][batch_index]  # Corresponding labels
        aligned_labels.append(align_labels_with_tokens(labels, word_ids))

    # Add aligned labels to tokenized inputs
    tokenized_inputs["labels"] = aligned_labels

    return tokenized_inputs
```

# Fine tuning

```python
  from transformers import AutoTokenizer, AutoModelForTokenClassification, TrainingArguments, Trainer, DataCollatorForTokenClassification
  import os
  import matplotlib.pyplot as plt
  from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
  import numpy as np

  # Danh sách các mô hình tiền huấn luyện
  model_checkpoints = [
      "bert-base-multilingual-cased",
  ]

  # Tạo thư mục lưu kết quả
  output_base_dir = "./finetuned_models"
  os.makedirs(output_base_dir, exist_ok=True)

  # Lưu kết quả huấn luyện để vẽ biểu đồ
  results = []
  train_losses = []
  eval_losses = []
  train_accuracies = []
  eval_accuracies = []

  # Lặp qua từng mô hình trong danh sách
  for model_checkpoint in model_checkpoints:
      print(f"\nStarting fine-tuning for model: {model_checkpoint}\n")

      # Load tokenizer
      tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
      data_collator = DataCollatorForTokenClassification(tokenizer=tokenizer)

      # Tokenize và căn chỉnh nhãn
      tokenized_datasets = raw_datasets.map(
          lambda examples: tokenize_and_align_labels(examples),
          batched=True,
          remove_columns=raw_datasets["train"].column_names,
      )

      # Load model
      model = AutoModelForTokenClassification.from_pretrained(
          model_checkpoint,
          id2label=id2label,
          label2id=label2id,
      )

      def compute_metrics(p):
          predictions, labels = p
          predictions = np.argmax(predictions, axis=-1)
          true_labels = labels

          # Flatten the predictions and labels
          flat_preds = predictions.flatten()
          flat_labels = true_labels.flatten()

          # Calculate precision, recall, f1-score and accuracy
          precision = precision_score(flat_labels, flat_preds, average='weighted', zero_division=1)
          recall = recall_score(flat_labels, flat_preds, average='weighted', zero_division=1)
          f1 = f1_score(flat_labels, flat_preds, average='weighted', zero_division=1)
          accuracy = accuracy_score(flat_labels, flat_preds)

          return {"precision": precision, "recall": recall, "f1": f1, "accuracy": accuracy}

      # Huấn luyện mô hình
      training_args = TrainingArguments(
          output_dir=f"{output_base_dir}/{model_checkpoint.replace('/', '-')}",
          evaluation_strategy="epoch",
          logging_dir="./logs",
          logging_steps=10,
          save_steps=10,
          num_train_epochs=3,
      )

      # Trainer
      trainer = Trainer(
          model=model,
          args=training_args,
          train_dataset=tokenized_datasets["train"],
          eval_dataset=tokenized_datasets["validation"],
          data_collator=data_collator,
          compute_metrics=compute_metrics,
      )

      # Huấn luyện mô hình và lấy kết quả
      train_result = trainer.train()

      # Lưu mô hình và tokenizer
      trainer.save_model(f"{output_base_dir}/{model_checkpoint.replace('/', '-')}")
      tokenizer.save_pretrained(f"{output_base_dir}/{model_checkpoint.replace('/', '-')}")
      print(f"Model {model_checkpoint} fine-tuned and saved.\n")

      # Lưu kết quả huấn luyện
      metrics = train_result.metrics
      results.append({
          "model": model_checkpoint,
          "train_loss": metrics["train_loss"],
          "eval_loss": metrics.get("eval_loss", None),
          "train_accuracy": metrics.get("train_accuracy", None),
          "eval_accuracy": metrics.get("eval_accuracy", None),
          "precision": metrics.get("eval_precision", None),
          "recall": metrics.get("eval_recall", None),
          "f1": metrics.get("eval_f1", None),
          "accuracy": metrics.get("eval_accuracy", None)
      })

      # Lưu loss và accuracy
      train_losses.append(metrics["train_loss"])
      eval_losses.append(metrics.get("eval_loss", None))
      train_accuracies.append(metrics.get("train_accuracy", None))
      eval_accuracies.append(metrics.get("eval_accuracy", None))

```
