# **Lstm-to-Lstm without attention**

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.nn.utils.rnn import pad_sequence, pack_padded_sequence, pad_packed_sequence
from torch.utils.data import Dataset, DataLoader
from pyvi.ViTokenizer import tokenize
from nltk.tokenize import word_tokenize
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from collections import Counter
import nltk

nltk.download('punkt')

# Hyperparameters
EMBED_SIZE = 256
HIDDEN_SIZE = 512
BATCH_SIZE = 64
EPOCHS = 15
LEARNING_RATE = 0.001
MAX_LEN = 20
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
PAD_IDX = 0
SOS_IDX = 2  # Start of sentence index (will be added in vocab)
```

```python
# Tokenization Functions
def vi_tokenizer(sentence):
    return tokenize(sentence).split()

def en_tokenizer(sentence):
    return word_tokenize(sentence)

# Simulated dataset (for testing)
input_texts = input_text[:120000]
output_texts = output_text[:120000]

# Build Vocabulary
def build_vocab_optimized(texts, tokenizer, max_vocab_size=6000):
    counter = Counter(word for text in texts for word in tokenizer(text))
    most_common = counter.most_common(max_vocab_size - 2)
    vocab = {word: idx + 2 for idx, (word, _) in enumerate(most_common)}
    vocab["<PAD>"] = PAD_IDX
    vocab["<UNK>"] = 1
    vocab["<SOS>"] = SOS_IDX
    return vocab

input_vocab = build_vocab_optimized(input_texts, vi_tokenizer)
output_vocab = build_vocab_optimized(output_texts, en_tokenizer)
reverse_output_vocab = {idx: word for word, idx in output_vocab.items()}

# Tokenize and Pad Sequences
def tokenize_and_pad(texts, tokenizer, vocab, max_len=MAX_LEN):
    sequences = [[vocab.get(word, vocab["<UNK>"]) for word in tokenizer(text)] for text in texts]
    return pad_sequence([torch.tensor(seq[:max_len]) for seq in sequences], batch_first=True, padding_value=PAD_IDX)

input_seqs = tokenize_and_pad(input_texts, vi_tokenizer, input_vocab)
output_seqs = tokenize_and_pad(output_texts, en_tokenizer, output_vocab)

# Dataset and DataLoader
class TranslationDataset(Dataset):
    def __init__(self, inputs, targets):
        self.inputs = inputs
        self.targets = targets

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):
        return self.inputs[idx], self.targets[idx]

# Split data into training and testing sets
train_inputs, test_inputs, train_targets, test_targets = train_test_split(
    input_seqs, output_seqs, test_size=0.2, random_state=42
)

train_dataset = TranslationDataset(train_inputs, train_targets)
test_dataset = TranslationDataset(test_inputs, test_targets)

train_dataloader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, pin_memory=True)
test_dataloader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, pin_memory=True)

# Model Definitions
class Encoder(nn.Module):
    def __init__(self, input_vocab_size, embedding_dim, hidden_dim):
        super(Encoder, self).__init__()
        self.embedding = nn.Embedding(input_vocab_size, embedding_dim, padding_idx=PAD_IDX)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)

    def forward(self, x, lengths):
        embedded = self.embedding(x)
        lengths = lengths.cpu()
        packed_input = pack_padded_sequence(embedded, lengths, batch_first=True, enforce_sorted=False)
        packed_outputs, (hidden, cell) = self.lstm(packed_input)
        outputs, _ = pad_packed_sequence(packed_outputs, batch_first=True, padding_value=PAD_IDX)
        return outputs, (hidden, cell)

class Decoder(nn.Module):
    def __init__(self, output_vocab_size, embedding_dim, hidden_dim):
        super(Decoder, self).__init__()
        self.embedding = nn.Embedding(output_vocab_size, embedding_dim, padding_idx=PAD_IDX)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_vocab_size)

    def forward(self, x, hidden, cell):
        embedded = self.embedding(x).unsqueeze(1)
        output, (hidden, cell) = self.lstm(embedded, (hidden, cell))
        predictions = self.fc(output.squeeze(1))
        return predictions, hidden, cell

class Seq2Seq(nn.Module):
    def __init__(self, encoder, decoder, device):
        super(Seq2Seq, self).__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.device = device

    def forward(self, source, target, source_lengths, teacher_forcing_ratio=0.5):
        batch_size = source.size(0)
        target_len = target.size(1)
        target_vocab_size = len(output_vocab)

        outputs = torch.zeros(batch_size, target_len, target_vocab_size).to(self.device)
        encoder_outputs, (hidden, cell) = self.encoder(source, source_lengths)

        x = target[:, 0]
        for t in range(1, target_len):
            output, hidden, cell = self.decoder(x, hidden, cell)
            outputs[:, t] = output
            teacher_force = torch.rand(1).item() < teacher_forcing_ratio
            x = target[:, t] if teacher_force else output.argmax(1)
        return outputs

# Initialize models
encoder = Encoder(len(input_vocab), EMBED_SIZE, HIDDEN_SIZE).to(DEVICE)
decoder = Decoder(len(output_vocab), EMBED_SIZE, HIDDEN_SIZE).to(DEVICE)
seq2seq_no_attention = Seq2Seq(encoder, decoder, DEVICE).to(DEVICE)

# Loss and optimizer
criterion = nn.CrossEntropyLoss(ignore_index=PAD_IDX)
optimizer_no_attention = optim.Adam(seq2seq_no_attention.parameters(), lr=LEARNING_RATE)

# Training and Evaluation
def calculate_accuracy(model, dataloader):
    model.eval()
    total = 0
    correct = 0
    with torch.no_grad():
        for source, target in dataloader:
            source, target = source.to(DEVICE), target.to(DEVICE)
            lengths = (source != PAD_IDX).sum(dim=1)
            outputs = model(source, target, lengths, teacher_forcing_ratio=0)
            predicted = outputs.argmax(dim=2)
            for i in range(target.size(0)):
                target_seq = target[i][1:]
                predicted_seq = predicted[i][1:]
                target_seq = target_seq[target_seq != PAD_IDX]
                predicted_seq = predicted_seq[:len(target_seq)]
                total += len(target_seq)
                correct += (target_seq == predicted_seq).sum().item()
    return correct / total

import matplotlib.pyplot as plt

def train_model_with_accuracy_and_logging(model, optimizer, criterion, train_loader, test_loader, epochs=10):
    best_train_acc = 0
    best_test_acc = 0
    best_epoch = 0

    train_losses = []
    test_losses = []
    train_accuracies = []
    test_accuracies = []

    for epoch in range(epochs):
        model.train()
        total_loss = 0
        for source, target in train_loader:
            source, target = source.to(DEVICE), target.to(DEVICE)
            lengths = (source != PAD_IDX).sum(dim=1)
            optimizer.zero_grad()
            outputs = model(source, target, lengths)
            outputs = outputs[:, 1:].reshape(-1, outputs.shape[-1])
            targets = target[:, 1:].reshape(-1)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()

        avg_train_loss = total_loss / len(train_loader)
        train_losses.append(avg_train_loss)

        train_acc = calculate_accuracy(model, train_loader)
        test_acc = calculate_accuracy(model, test_loader)

        train_accuracies.append(train_acc)
        test_accuracies.append(test_acc)

        if test_acc > best_test_acc:
            best_train_acc = train_acc
            best_test_acc = test_acc
            best_epoch = epoch + 1

        print(f"Epoch {epoch + 1}/{epochs}, Loss: {avg_train_loss:.4f}, "
              f"Train Accuracy: {train_acc:.4f}, Test Accuracy: {test_acc:.4f}")

    print(f"Best Test Accuracy: {best_test_acc:.4f} at Epoch {best_epoch}")
    return train_losses, test_losses, train_accuracies, test_accuracies

# Train the models and log the metrics
print("Training model without attention: ")
train_losses_no_attention, _, train_acc_no_attention, test_acc_no_attention = train_model_with_accuracy_and_logging(
    seq2seq_no_attention, optimizer_no_attention, criterion, train_dataloader, test_dataloader, EPOCHS
)

# Plotting results
plt.figure(figsize=(12, 6))

# Plot training loss
plt.subplot(1, 2, 1)
plt.plot(range(1, EPOCHS + 1), train_losses_no_attention, label='Without Attention - Loss', marker='o')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss Comparison')
plt.legend()

# Plot accuracies
plt.subplot(1, 2, 2)
plt.plot(range(1, EPOCHS + 1), train_acc_no_attention, label='Without Attention - Train Acc', marker='o')
plt.plot(range(1, EPOCHS + 1), test_acc_no_attention, label='Without Attention - Test Acc', marker='o')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison')
plt.legend()

plt.tight_layout()
plt.show()
```

## **Evaluate BLEU score for model no attention**

```python
from nltk.translate.bleu_score import corpus_bleu, SmoothingFunction

# Tokenize predictions and references
def detokenize_and_convert_to_sentences(indices, vocab):
    sentences = []
    for seq in indices:
        words = [vocab[idx] for idx in seq if idx not in [PAD_IDX, SOS_IDX]]
        sentences.append(" ".join(words))
    return sentences

# Evaluate BLEU score for a given model
def evaluate_bleu_score(model, dataloader, vocab):
    model.eval()
    predictions = []
    references = []
    with torch.no_grad():
        for source, target in dataloader:
            source, target = source.to(DEVICE), target.to(DEVICE)
            lengths = (source != PAD_IDX).sum(dim=1)
            outputs = model(source, target, lengths, teacher_forcing_ratio=0)  # Disable teacher forcing
            predicted_indices = outputs.argmax(dim=2).cpu().numpy()
            target_indices = target.cpu().numpy()

            # Convert to sentences
            predicted_sentences = detokenize_and_convert_to_sentences(predicted_indices, vocab)
            reference_sentences = detokenize_and_convert_to_sentences(target_indices, vocab)

            # Append results
            predictions.extend(predicted_sentences)
            references.extend([[ref] for ref in reference_sentences])  # BLEU expects list of lists

    # Calculate BLEU score
    smoothing_function = SmoothingFunction().method1
    bleu_score = corpus_bleu(references, predictions, smoothing_function=smoothing_function)
    return bleu_score

# Evaluate BLEU score for model no attention
print("Calculating BLEU score for model without Attention:")
bleu_no_attention = evaluate_bleu_score(seq2seq_no_attention, test_dataloader, reverse_output_vocab)
print(f"BLEU score (No Attention): {bleu_no_attention:.4f}")
```

## **Prediction**

```python

def batch_predict(model, sentences, tokenizer, vocab, reverse_vocab, max_len=MAX_LEN):
    model.eval()
    predictions = []
    with torch.no_grad():
        for sentence in sentences:
            tokens = [vocab.get(word, vocab["<UNK>"]) for word in tokenizer(sentence)]
            input_tensor = torch.tensor(tokens).unsqueeze(0).to(DEVICE)
            input_lengths = torch.tensor([len(tokens)]).to(DEVICE)
            encoder_outputs, (hidden, cell) = model.encoder(input_tensor, input_lengths)
            x = torch.tensor([vocab["<SOS>"]]).to(DEVICE)
            translated_sentence = []
            for _ in range(max_len):
                output, hidden, cell = model.decoder(x, hidden, cell)
                predicted_token = output.argmax(1).item()
                if predicted_token == PAD_IDX:  # Kết thúc nếu gặp PAD
                    break
                if predicted_token != SOS_IDX:  # Bỏ qua token <SOS>
                    translated_sentence.append(reverse_vocab[predicted_token])
                x = torch.tensor([predicted_token]).to(DEVICE)
            predictions.append(" ".join(translated_sentence))
    return predictions
```

# **Lstm-to-Lstm with attention**

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.nn.utils.rnn import pad_sequence, pack_padded_sequence, pad_packed_sequence
from torch.utils.data import Dataset, DataLoader
from pyvi.ViTokenizer import tokenize
from nltk.tokenize import word_tokenize
import nltk
from collections import Counter
from sklearn.model_selection import train_test_split
import pandas as pd

# Đọc file CSV
file_path = "/content/vietnamese_english.csv"
df = pd.read_csv(file_path)

# Lấy tất cả các câu hỏi tiếng Việt và tiếng Anh vào hai danh sách
input_text = df['Vietnamese'].tolist()
output_text = df['English'].tolist()
nltk.download('punkt')

# Hyperparameters
EMBED_SIZE = 256
HIDDEN_SIZE = 512
BATCH_SIZE = 64
EPOCHS = 25
LEARNING_RATE = 0.001
MAX_LEN = 20
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
PAD_IDX = 0
SOS_IDX = 2  # Start of sentence index

# Tokenization Functions
def vi_tokenizer(sentence):
    return tokenize(sentence).split()

def en_tokenizer(sentence):
    return word_tokenize(sentence)

# Simulated dataset (for testing)
input_texts = input_text[:120000]
output_texts = output_text[:120000]

# Build Vocabulary
def build_vocab_optimized(texts, tokenizer, max_vocab_size=6000):
    counter = Counter(word for text in texts for word in tokenizer(text))
    most_common = counter.most_common(max_vocab_size - 2)
    vocab = {word: idx + 2 for idx, (word, _) in enumerate(most_common)}
    vocab["<PAD>"] = PAD_IDX
    vocab["<UNK>"] = 1
    vocab["<SOS>"] = SOS_IDX
    return vocab

input_vocab = build_vocab_optimized(input_texts, vi_tokenizer)
output_vocab = build_vocab_optimized(output_texts, en_tokenizer)
reverse_output_vocab = {idx: word for word, idx in output_vocab.items()}

# Tokenize and Pad Sequences
def tokenize_and_pad(texts, tokenizer, vocab, max_len=MAX_LEN):
    sequences = [[vocab.get(word, vocab["<UNK>"]) for word in tokenizer(text)] for text in texts]
    return pad_sequence([torch.tensor(seq[:max_len]) for seq in sequences], batch_first=True, padding_value=PAD_IDX)

input_seqs = tokenize_and_pad(input_texts, vi_tokenizer, input_vocab)
output_seqs = tokenize_and_pad(output_texts, en_tokenizer, output_vocab)

# Dataset Class
class TranslationDataset(Dataset):
    def __init__(self, inputs, targets):
        self.inputs = inputs
        self.targets = targets

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):
        return self.inputs[idx], self.targets[idx]

# Split data into training and testing sets
train_inputs, test_inputs, train_targets, test_targets = train_test_split(
    input_seqs, output_seqs, test_size=0.2, random_state=42
)

train_dataset = TranslationDataset(train_inputs, train_targets)
test_dataset = TranslationDataset(test_inputs, test_targets)

train_dataloader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, pin_memory=True)
test_dataloader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, pin_memory=True)

# Model Definitions
class Encoder(nn.Module):
    def __init__(self, input_vocab_size, embedding_dim, hidden_dim):
        super(Encoder, self).__init__()
        self.embedding = nn.Embedding(input_vocab_size, embedding_dim, padding_idx=PAD_IDX)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)

    def forward(self, x, lengths):
        embedded = self.embedding(x)
        lengths = lengths.cpu()
        packed_input = pack_padded_sequence(embedded, lengths, batch_first=True, enforce_sorted=False)
        packed_outputs, (hidden, cell) = self.lstm(packed_input)
        outputs, _ = pad_packed_sequence(packed_outputs, batch_first=True, padding_value=PAD_IDX)
        return outputs, (hidden, cell)

class DecoderWithAttention(nn.Module):
    def __init__(self, output_vocab_size, embedding_dim, hidden_dim):
        super(DecoderWithAttention, self).__init__()
        self.embedding = nn.Embedding(output_vocab_size, embedding_dim, padding_idx=PAD_IDX)
        self.attn = nn.Linear(hidden_dim * 2, MAX_LEN)
        self.attn_combine = nn.Linear(hidden_dim + embedding_dim, hidden_dim)
        self.lstm = nn.LSTM(hidden_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_vocab_size)

    def forward(self, x, encoder_outputs, hidden, cell):
        seq_len = encoder_outputs.size(1)
        embedded = self.embedding(x).unsqueeze(1)
        attn_weights = torch.softmax(
            self.attn(torch.cat((hidden[-1], encoder_outputs[:, -1]), dim=1)), dim=1
        )[:, :seq_len]
        context = torch.bmm(attn_weights.unsqueeze(1), encoder_outputs)
        lstm_input = torch.cat((embedded, context), dim=2)
        lstm_input = self.attn_combine(lstm_input)
        output, (hidden, cell) = self.lstm(lstm_input, (hidden, cell))
        predictions = self.fc(output.squeeze(1))
        return predictions, hidden, cell

class Seq2Seq(nn.Module):
    def __init__(self, encoder, decoder, device):
        super(Seq2Seq, self).__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.device = device

    def forward(self, source, target, source_lengths, teacher_forcing_ratio=0.5):
        batch_size = source.size(0)
        target_len = target.size(1)
        target_vocab_size = len(output_vocab)

        outputs = torch.zeros(batch_size, target_len, target_vocab_size).to(self.device)
        encoder_outputs, (hidden, cell) = self.encoder(source, source_lengths)

        x = target[:, 0]
        for t in range(1, target_len):
            output, hidden, cell = self.decoder(x, encoder_outputs, hidden, cell)
            outputs[:, t] = output
            teacher_force = torch.rand(1).item() < teacher_forcing_ratio
            x = target[:, t] if teacher_force else output.argmax(1)
        return outputs

# Initialize models
encoder = Encoder(len(input_vocab), EMBED_SIZE, HIDDEN_SIZE).to(DEVICE)
decoder_with_attention = DecoderWithAttention(len(output_vocab), EMBED_SIZE, HIDDEN_SIZE).to(DEVICE)
seq2seq_with_attention = Seq2Seq(encoder, decoder_with_attention, DEVICE).to(DEVICE)

# Loss and optimizer
criterion = nn.CrossEntropyLoss(ignore_index=PAD_IDX)
optimizer_with_attention = optim.Adam(seq2seq_with_attention.parameters(), lr=LEARNING_RATE)

# Accuracy Calculation
def calculate_accuracy(model, dataloader):
    model.eval()
    total = 0
    correct = 0
    with torch.no_grad():
        for source, target in dataloader:
            source, target = source.to(DEVICE), target.to(DEVICE)
            lengths = (source != PAD_IDX).sum(dim=1)
            outputs = model(source, target, lengths, teacher_forcing_ratio=0)
            predicted = outputs.argmax(dim=2)
            for i in range(target.size(0)):
                target_seq = target[i][1:]
                predicted_seq = predicted[i][1:]
                target_seq = target_seq[target_seq != PAD_IDX]
                predicted_seq = predicted_seq[:len(target_seq)]
                total += len(target_seq)
                correct += (target_seq == predicted_seq).sum().item()
    return correct / total

# Train Model
import matplotlib.pyplot as plt

def train_model_with_accuracy_and_logging(model, optimizer, criterion, train_loader, test_loader, epochs=10):
    best_train_acc = 0
    best_test_acc = 0
    best_epoch = 0

    train_losses = []
    test_losses = []
    train_accuracies = []
    test_accuracies = []

    for epoch in range(epochs):
        model.train()
        total_loss = 0
        for source, target in train_loader:
            source, target = source.to(DEVICE), target.to(DEVICE)
            lengths = (source != PAD_IDX).sum(dim=1)
            optimizer.zero_grad()
            outputs = model(source, target, lengths)
            outputs = outputs[:, 1:].reshape(-1, outputs.shape[-1])
            targets = target[:, 1:].reshape(-1)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()

        avg_train_loss = total_loss / len(train_loader)
        train_losses.append(avg_train_loss)

        train_acc = calculate_accuracy(model, train_loader)
        test_acc = calculate_accuracy(model, test_loader)

        train_accuracies.append(train_acc)
        test_accuracies.append(test_acc)

        if test_acc > best_test_acc:
            best_train_acc = train_acc
            best_test_acc = test_acc
            best_epoch = epoch + 1

        print(f"Epoch {epoch + 1}/{epochs}, Loss: {avg_train_loss:.4f}, "
              f"Train Accuracy: {train_acc:.4f}, Test Accuracy: {test_acc:.4f}")

    print(f"Best Test Accuracy: {best_test_acc:.4f} at Epoch {best_epoch}")
    return train_losses, test_losses, train_accuracies, test_accuracies

# Train the models and log the metrics
print("Training model with attention: ")
train_losses_with_attention, _, train_acc_with_attention, test_acc_with_attention = train_model_with_accuracy_and_logging(
    seq2seq_with_attention, optimizer_with_attention, criterion, train_dataloader, test_dataloader, EPOCHS
)

# Plotting results
plt.figure(figsize=(12, 6))

# Plot training loss
plt.subplot(1, 2, 1)
plt.plot(range(1, EPOCHS + 1), train_losses_with_attention, label='With Attention - Loss', marker='o')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss Comparison')
plt.legend()

# Plot accuracies
plt.subplot(1, 2, 2)
plt.plot(range(1, EPOCHS + 1), train_acc_with_attention, label='With Attention - Train Acc', marker='o')
plt.plot(range(1, EPOCHS + 1), test_acc_with_attention, label='With Attention - Test Acc', marker='o')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison')
plt.legend()

plt.tight_layout()
plt.show()
```

## **Evaluate BLEU score for model with Attention**

```python
from nltk.translate.bleu_score import corpus_bleu, SmoothingFunction

# Tokenize predictions and references
def detokenize_and_convert_to_sentences(indices, vocab):
    sentences = []
    for seq in indices:
        words = [vocab[idx] for idx in seq if idx not in [PAD_IDX, SOS_IDX]]
        sentences.append(" ".join(words))
    return sentences

# Evaluate BLEU score for a given model
def evaluate_bleu_score(model, dataloader, vocab):
    model.eval()
    predictions = []
    references = []
    with torch.no_grad():
        for source, target in dataloader:
            source, target = source.to(DEVICE), target.to(DEVICE)
            lengths = (source != PAD_IDX).sum(dim=1)
            outputs = model(source, target, lengths, teacher_forcing_ratio=0)  # Disable teacher forcing
            predicted_indices = outputs.argmax(dim=2).cpu().numpy()
            target_indices = target.cpu().numpy()

            # Convert to sentences
            predicted_sentences = detokenize_and_convert_to_sentences(predicted_indices, vocab)
            reference_sentences = detokenize_and_convert_to_sentences(target_indices, vocab)

            # Append results
            predictions.extend(predicted_sentences)
            references.extend([[ref] for ref in reference_sentences])  # BLEU expects list of lists

    # Calculate BLEU score
    smoothing_function = SmoothingFunction().method1
    bleu_score = corpus_bleu(references, predictions, smoothing_function=smoothing_function)
    return bleu_score

# Evaluate BLEU score for model with Attention
print("Calculating BLEU score for model with Attention:")
bleu_with_attention = evaluate_bleu_score(seq2seq_with_attention, test_dataloader, reverse_output_vocab)
print(f"BLEU score (With Attention): {bleu_with_attention:.4f}")
```

## **Prediction**

```python
def predict_batch(model, inputs, vocab, reverse_vocab, max_len=MAX_LEN):
    model.eval()
    predictions = []
    with torch.no_grad():
        for input_seq in inputs:
            input_seq = input_seq.unsqueeze(0).to(DEVICE)  # Thêm batch dimension
            input_lengths = torch.tensor([len(input_seq[input_seq != PAD_IDX])]).to(DEVICE)
            encoder_outputs, (hidden, cell) = model.encoder(input_seq, input_lengths)
            x = torch.tensor([SOS_IDX]).to(DEVICE)  # Khởi tạo với token <SOS>
            translated_sentence = []
            for _ in range(max_len):
                output, hidden, cell = model.decoder(x, encoder_outputs, hidden, cell)
                predicted_token = output.argmax(1).item()
                if predicted_token == PAD_IDX:  # Dừng khi gặp token PAD
                    break
                if predicted_token != SOS_IDX:  # Không thêm token <SOS> vào kết quả
                    translated_sentence.append(reverse_vocab.get(predicted_token, "<UNK>"))
                x = torch.tensor([predicted_token]).to(DEVICE)
            predictions.append(" ".join(translated_sentence))
    return predictions
```

```python
def predict_single_sample(model, sample, sample_length, target_vocab, max_len=MAX_LEN):
    model.eval()
    with torch.no_grad():
        sample = sample.unsqueeze(0).to(DEVICE)  # Add batch dimension
        sample_length = torch.tensor([sample_length]).to(DEVICE)  # Add batch dimension
        encoder_outputs, (hidden, cell) = model.encoder(sample, sample_length)

        # Start decoding
        output_sentence = []
        x = torch.tensor([SOS_IDX]).to(DEVICE)  # Start token

        for _ in range(max_len):
            # Ensure both hidden and cell are passed for attention model
            if hasattr(model.decoder, 'attn'):  # Check if model has attention mechanism
                # Pass encoder_outputs, hidden, and cell for the attention-based decoder
                output, hidden, cell = model.decoder(x, encoder_outputs, hidden, cell)
            else:
                # If no attention mechanism, just pass hidden and cell
                output, hidden, cell = model.decoder(x, hidden, cell)

            predicted_idx = output.argmax(1).item()
            if predicted_idx == PAD_IDX:
                break
            output_sentence.append(reverse_output_vocab.get(predicted_idx, "<UNK>"))
            x = torch.tensor([predicted_idx]).to(DEVICE)

    return " ".join(output_sentence)



# Prepare sample from train data
sample_input = train_inputs[7]
sample_length = (sample_input != PAD_IDX).sum().item()
sample_target = train_targets[7]
sample_target_text = " ".join(reverse_output_vocab[idx.item()] for idx in sample_target if idx != PAD_IDX)

# Predict using both models
predicted_no_attention = predict_single_sample(seq2seq_no_attention, sample_input, sample_length, reverse_output_vocab)
predicted_with_attention = predict_single_sample(seq2seq_with_attention, sample_input, sample_length, reverse_output_vocab)

print("\n--- Single Sample Prediction ---\n")
# print("Input Sentence:", " ".join(input_vocab.keys()))
print("Target Sentence:", sample_target_text)
print("Predicted (Without Attention):", predicted_no_attention)
print("Predicted (With Attention):", predicted_with_attention)
```
